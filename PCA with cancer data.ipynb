```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
print(cancer.keys())
```

```python
print(cancer['DESCR'])
```

```python
## making a dataframe
```

```python
df = pd.DataFrame(cancer['data'], columns =cancer['feature_names'])
```

```python
df.head()
```

```python
cancer['target_names']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

# PCA 

```python
scaled_data.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
# hence we have decomposed data scaled data into 2 components which was initially in 30 components or features 
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=x_pca[:,0],y= x_pca[:,1], data= x_pca, c = cancer['target'],cmap='rainbow')
plt.xlabel("First principle component ")
plt.ylabel('Second principle component')
```

```python
 pca.components_
```

```python
df_compo = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
df_compo
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df_compo)
```
