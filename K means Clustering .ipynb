```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
# Creating a artificial dataset from sklearn 
```

```python
data = make_blobs(n_samples=200, n_features=2, centers = 5, cluster_std=1.5, random_state= 101)
```

```python
data[1]
```

```python
plt.grid(True)
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
# Using sklearn for clustering algorit
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=8)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
# kmeans.labels_ is our predicted data and data[1] is our original data 
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,sharey = True, figsize=(10,6))

ax1.set_title(" K means ")
ax1.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_, cmap='rainbow')


ax2.set_title(" Original ")
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap='rainbow')
```
